Importando o opencv e lendo a imagem

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('imagens/lena.jpg', 0)
```

```python
pimage = image.copy()
pyramid = [pimage]
for i in range(0,6):
    pimage = cv2.pyrDown(pimage)
    pyramid.append(pimage)
    plt.imshow(pimage, aspect='auto', cmap='gray')
    plt.show()
```
