# Histograma da imagem

Importando as dependências

```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

Importando a imagem colorida e outra em escala de cinza

```python

image_gray = cv2.imread('imagens/lena.jpg', 0)
```

Gerando e plotando o histograma para escala de cinza

```python
hist_full = cv2.calcHist([image_gray], [0], None, [256], [0, 256])
plt.subplot(221), plt.imshow(image_gray, 'gray')
plt.subplot(222), plt.plot(hist_full)
plt.xlim([0, 256])
plt.show()
```

Histograma para imagem rgb

```python
image_color = cv2.imread('imagens/lena.jpg')
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([image_color], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
plt.show()
plt.imshow(image_color, cmap='brg')
plt.show()
```
